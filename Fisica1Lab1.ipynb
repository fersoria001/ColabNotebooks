##El momento lineal

Para expresar el momento lineal recordemos la segunda ley de Newton.

Definidas las siguientes cantidades:
* $\vec F$ = vector fuerza
* $m$ = masa
* $\vec a$ = aceleración

La segunda ley de Newton dice lo siguiente.

$\sum \vec F = m \cdot \vec a$

Sea $m$ una particula de masa constante, $\vec v$ una velocidad y $t$ un tiempo.
Como $ \vec a  = \frac{d \vec v}{dt}$ podemos escribir la segunda ley de Newton asi:
$\sum(F) = m \frac{d\vec v}{dt} = \frac{{d}}{{dt}}(m \cdot v)$
Introducimos $m$ en la derivada por la regla de la constante, esto quiere decir que la sumatoria de fuerzas que actua sobre una particula es igual a la rapidez de cambio del producto de la masa y la velocidad. Llamamos a este producto -momento lineal- y se simboliza con el vector $\vec p$.
* $\vec p$ = momento lineal
* $m$ = masa
* $\vec v$ = velocidad
* $\vec p$ = $m \cdot \vec v$ momento lineal

El momento lineal es uan cantidad vectorial con la misma direccion que la velocidad de la particula.
Como es un vector, es de utilidad expresar el momento lineal en funcion de sus componentes
* $p_x = m \cdot \vec v_x$ 
* $p_y = m \cdot \vec v_y$
* $p_z = m \cdot \vec v_z$

Las unidades para el momento lineal en el sistema internacional son $kg \cdot \frac {m}{s}$.
Como el momento lineal deriva de la segunda ley de newton y la sumatoria de fuerzas tambien es igual a la rapidez de cambio del producto de la masa y la velocidad a la que llamamos p-momento lineal- podemos sustituir en la ecuacion de la segunda ley de Newton:
* $\sum \vec F = m \cdot \vec a$
* por
* $\sum \vec F = \frac {d\vec p}{dt} $

Teorema del impulso y el momento lineal:

Tanto el  momento lineal $\vec p=m \cdot \vec v$ y su energia cinetica $K = \frac{1}{2}
m\vec v^2  $ dependen de la masa y la velocidad de la particula. La principal diferencia es que el momento lineal es un vector y la energia cinetica es un escalar. Sin embargo para ver la diferencia fisica entre ambos debemos definir una cantidad relacionada con el momento: el impulso.

Si tenemos una particula sobre la que actua una fuerza constante $\sum \vec F$
en un intervalo de tiempo $\Delta t = t2-t1.$

El impulso de la fuerza neta denotado con $J$ es el producto de la fuerza neta y el intervalo de tiempo. 
$J = sum(F) * (t2-t1) = \sum \vec F \cdot \Delta t$
Es una cantidad vectorial, su direccion es la de la fuerza neta $\vec Fn$ de $\sum \vec F$ y su magnitud es el producto de la magnitud de la $\vec Fn$ y el tiempo en el que actua. Las unidades de $J$ son $ newton \cdot segundo$ o 
$kg \cdot m/s$ que son iguales a las de momento lineal.

Como habiamos planteado la segunda ley de Newton como la rapidez de cambio de $\vec p$.
* $\sum \vec F = \frac{d\vec p}{dt}$

Si la $Fn = \sum \vec F$ es constante, tambien lo es el momento, en tal caso
$\frac {d\vec p}{dt}$ es igual al cambio total de momento lineal $\vec p2- \vec p1$ durante el lapso $t2-t1$.

* $\sum \vec F = \frac {\vec p2- \vec p1}{t2-t1}$

y como

* $J = \sum \vec F \cdot (t2-t1) = \sum \vec F \cdot \Delta t$

Podemos compararlas y obtener que 

* $J = \vec p2- \vec p1$  (Este es el teorema del impulso y el momento lineal.)

## Energia cinetica:
Para analizar las colisiones tambien utilizaremos el teorema de conservacion de la Energia cinetica $K$. Como vimos anteriormente el teorema del impuslo y el momento lineal dice que los cambios en el momento lineal de una particular se deben al impulso, que depende del tiempo durante el que actua la fuerza neta,
en cambio el teorema de trabajo energia nos dice que la energia cinetica cambia cuando se realiza trabajo sobre una particula.

### El teorema del trabajo energia:
El teorema de trabajo energia expresado como:
* $Wtot = K_2 - K_1$

Nos dice que la energia cinetica cambia cuando se realiza trabajo sobre una particular, el trabajo total depende de la distancia con la que actua la fuerza neta.

Considere una particula que parte del reposo en $t_1$, de manera que $\vec v_1 = 0$ y su energia cinetica inicial $K_1 = \frac{1}{2}mv_1^2 = 0$.

Ahora una fuerza neta constante $\vec F$ actua sobre el cuerpo del tiempo $t_1$ al tiempo $t_2$. En este intervalo de tiempo, la particula se mueve en una distancia $s$ en la direccion de la fuerza $\vec F$
Si utilizamos el teorema del impulso y el momento lineal y despejamos $\vec p_2$
* $J = \vec p2- \vec p1$
* $\vec p_2 = \vec p1 + \vec J$

En donde $\vec J = \vec F(t_2 - t_1)$ es el impulso que actua sobre la particula. Asi, el momento lineal de una particula es igual al impulso que la acelero desde el reposo hasta su rapidez actual. el impulso es el producto de la fuerza neta que acelero el cuerpo y el tiempo requerido para la aceleracion.
En cambio la energica cinetica $K$ del cuerpo en $t_2$ es
$K_2 = Wtot = Fs$, el trabajo total efectuado sobre el cuerpo para acelerarlo desde $t_1$.

## Conservacion del momento lineal
Para saber porque el momento lineal se conserva cuando dos cuerpos interactuan consideremos un sistema ideal de dos cuerpos que interactuan entre si. Consideremos que cada cuerpo ejerce una fuerza sobre el otro segun la tercera ley de Newton(accion y reaccion). Las fuerza del cuerpo A que realiza sobre el cuerpo B es igual a la fuerza que ejerce el cuerpo B sobre el cuerpo A pero en direccion contraria.

En cualquier sistema, las fuerzas que los cuerpos del sistema ejercen entre si se denominan fuerzas internas, las ejercidas por un cuerpo externo sobre cualquier cuerpo del sistema se denominan fuerzas externas.

Como no vamos a tomar en cuenta las fuerzas externas para este estudio, ya que es una situacion ideal en la que solo existen fuerzas internas, nuestro sistema es un sistema aislado. Las razones de cambio del momento lineal de ambos cuerpos son

$$ \vec F_B \text{ sobre A } = \frac{d\vec p_A}{dt} = \vec F_A \text{ sobre B } = \frac{d\vec p_B}{dt} $$

Como son iguales pero de direccion contraria su sumamos $\vec F_B + \vec F_A = 0$ asi que la razon de cambio de la suma vectorial $\vec p_A + \vec p_B$ es cero. 
Ahora definimos el momento lineal total $\vec P$ del sistema como esta suma vectorial, esto es
$$ \vec P = \vec p_A + \vec p_B$$
La razon de cambio del momento lineal total $\vec P$ es cero. Por lo tanto el momento lineal total del sistema es constante, aunque los momentos lineales individuales de los cuerpos que constituyen este sistema aislado pueden cambiar.

## Conservacion del momento lineal y choques
Con choques nos vamos a referir a todo impacto de un cuerpo sobre otro, los analizaremos en un sistema aislado ideal, porque asumimos que las fuerzas entre los cuerpos son mucho mayores que las externas. Entonces el momento lineal se conserva y es igual en el momento inicial $t_1$  y final $t_2$.

#### Choques elasticos e inelasticos
Si las fuerzas entre los cuerpos son conservativas, la energia cinetica total del sistema es la misma antes y despues del choque. Esto se denomina choque elastico. Un choque en el que la energia cinetica total final es menor que la inicial es un choque inelastico.
#### Choque totalmente inelastico
Es un choque en el que los cuerpos quedan pegados despues del choque,tienen la misma velocidad final $\vec v_2$:

La conservacion del momento lineal da la relacion:
$$\vec p = m \cdot \vec v$$
$$ m_A\vec v_{A1} + m_B\vec v_{B1} = (m_A + m_B)\vec v_2 $$

Si conocemos las masas y las velocidad iniciales, podremos calcular la velocidad final comun $\vec v_2$.
Como en el estudio a continuacion solo trabajaremos en un componente de movimiento $x$ uponga que un cuerpo con masa $m_A$ y componente $x$ inicial de velocidad $v_{A1x}$ choca inelasticamente con un cuerpo de masa $m_B$ en reposo  ($v_{B1x} = 0$) la componente x de velocidad despues del choque $v_{2x}$ es
$$ v_{2x} = \frac {m_A}{m_A + m_B} v_{A1x}$$

Como se trata de un choque inelastico la energia cinetica total final es menor que la inicial
$$ K_1 = \frac{1}{2}m_A v_{A1x}^2 $$
$$ K_2 = \frac{1}{2}(m_A + m_B) v_{2x}^2 = \frac{1}{2}(m_A + m_B) (\frac {m_A}{m_A + m_B})^2 v_{A1x}^2$$

El cociente de las energias cinetica final e inicial es:
$$ \frac {K_2}{K_1} = \frac {m_A}{m_A + m_B} $$ 
en un choque totalmente inelastico con B en reposo.

```python
#@title

datos = pd.read_csv('teo y fernando.csv', sep = ';')
datos.dropna(inplace=True)    
cols = datos.columns
#datos iniciales de las masas
masa_azul = 250.0
masa_rojo = 253.1
masa_pesa = 253.3
#elimino columnas innecesarias fecha,aceleracion y posicion
new_cols = [x for x in cols if not 'celerac' in x]
new_cols = [x for x in new_cols if not 'echa' in x]
new_cols = [x for x in new_cols if not 'osic' in x]
#reeplazo comas por puntos y convierto a float
for cn in datos.columns:
    if not cn.startswith('Fecha'):
        datos[cn] = datos[cn].str.replace(',', '.').astype(float)
#obtengo conjunto de tiempo,velocidad_carro1 y velocidad_carro2
#para el inicio y final de cada colision
# .loc() es para filas e iloc() para columnas
inicio_colision_elastico_MM = datos.loc[120, new_cols].iloc[:3]
final_colision_elastico_MM = datos.loc[126, new_cols].iloc[:3]
inicio_colision_inelastico_MM = datos.loc[78, new_cols].iloc[3:6].multiply(-1)
final_colision_inelastico_MM = datos.loc[84, new_cols].iloc[3:6].multiply(-1)
inicio_colision_elastico_rojo = datos.loc[66, new_cols].iloc[6:9]
final_colision_elastico_rojo = datos.loc[72, new_cols].iloc[6:9]
inicio_colision_inelastico_rojo = datos.loc[66, new_cols].iloc[9:12].multiply(-1) #revisar los datos
final_colision_inelastico_rojo = datos.loc[72, new_cols].iloc[9:12].multiply(-1)
inicio_colision_elastico_azul = datos.loc[102, new_cols].iloc[12:15]
final_colision_elastico_azul = datos.loc[108, new_cols].iloc[12:15] # o 108
inicio_colision_totalmente_inelastico_azul = datos.loc[54, new_cols].iloc[15:18].multiply(-1)
final_colision_totalmente_inelastico_azul = datos.loc[60, new_cols].iloc[15:18].multiply(-1)
#i = 0 = tiempo, i = 1 = velocidad azul, i = 2 = velocidad rojo
lista_colisiones = [
    inicio_colision_elastico_MM,
    final_colision_elastico_MM,
    inicio_colision_inelastico_MM,
    final_colision_inelastico_MM,
    inicio_colision_elastico_rojo,
    final_colision_elastico_rojo,
    inicio_colision_inelastico_rojo,
    final_colision_inelastico_rojo,
    inicio_colision_elastico_azul,
    final_colision_elastico_azul,
    inicio_colision_totalmente_inelastico_azul,
    final_colision_totalmente_inelastico_azul
]

#La conservacion del momento lineal establece que la suma de los momentos lineales
#antes de la colision debe ser igual a la suma de los momentos lineales
#despues de la colision
"""
Esta funcion calcula un momento lineal
@param masa float
@param velocidad float
@return masa * velocidad, float
"""
def calcular_momento_lineal(masa, velocidad):
    return masa * velocidad
"""
Esta funcion calcula la energia cinetica
@param masa, float
@param velocidad,float
@return energia_cinetica, float
"""
def calcular_energia_cinetica(masa,velocidad):
    energia_cinetica = (masa/2) * (velocidad**2)
    return energia_cinetica
#totalmente inelastico mazul > mrojo momento inicial
"""
Esta funcion calcula la suma de momentos y energia 
cinetica de ambos carros en base a la velocidad,
velocidad y masas del primer y segundo carro.
El nombre es extraido de las ultimas 2 palabras
del nombre de la primer columna.
@param colision list[tiempo,velocidadcarro1,velocidadcarro2]
@param masa1 = masa del primero carro, default 1
@param masa2 = masa del segundo carro, default 1
@param masa_pesa, se le suma a masa 1, default 0
@return list[suma_ek, suma_p, nombre_colision]
"""
def calcular_colision(colision,masa1=1,masa2=1,masa_pesa = 0):
  columna = colision.index[0]
  nombre = ''.join(columna.split()[-2:])
  vi_carro_uno = colision.iloc[[1]].values[0]
  vi_carro_dos = colision.iloc[[2]].values[0]
  if nombre.endswith("l"):
    momento_uno = calcular_momento_lineal(masa1,vi_carro_uno)
    ek_uno = calcular_energia_cinetica(masa1, vi_carro_uno)
    momento_dos = calcular_momento_lineal(masa2 + masa_pesa, vi_carro_dos)
    ek_dos = calcular_energia_cinetica(masa2 + masa_pesa, vi_carro_dos)
    descripcion = 'masa2 + masa_pesa'
  else:
    momento_uno = calcular_momento_lineal(masa1 + masa_pesa,vi_carro_uno)
    ek_uno = calcular_energia_cinetica(masa1 + masa_pesa, vi_carro_uno)
    momento_dos = calcular_momento_lineal(masa2, vi_carro_dos)
    ek_dos = calcular_energia_cinetica(masa2, vi_carro_dos)
    descripcion = 'masa1 + masa_pesa'
  ek_suma = ek_uno + ek_dos
  suma_momentos = momento_uno + momento_dos
  tiempo = colision.iloc[[0]].values[0]
  if tiempo < 0:
    tiempo = tiempo * -1
  return [nombre, tiempo , masa_pesa, ek_uno,ek_dos,momento_uno, momento_dos, descripcion]

##calculo las colisiones, en mi lista i < 4 son las MM (masas iguales)
resultados = []
for i in range(len(lista_colisiones)):
  if i >= 4:
    resultados.append(calcular_colision(lista_colisiones[i],masa_azul,masa_rojo,masa_pesa))
  else:
    resultados.append(calcular_colision(lista_colisiones[i],masa_azul,masa_rojo))
#print de resultados

import csv
csv_file = 'output.csv'
columns = []
column_data = {}
row_names = ['tiempo(s)', 'masa pesa(kg)', 'EK uno(J)', 'EK dos(J)', 'p_uno(J)', 'p_dos(J)', 'descripcion']
for i in range(len(resultados)):
    if i % 2 == 0:
        col_name = f'inicio_{resultados[i][0]}'
    else:
        col_name = f'final_{resultados[i][0]}'
    row_values = resultados[i][1:]
    column_data[col_name] = row_values
datos_post = pd.DataFrame(column_data, index=row_names)
datos_post.to_csv(csv_file)
print(resultados)
```
