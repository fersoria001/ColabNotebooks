```python

class Exercise:
    def __init__(self, name, sets, reps, rm, category, subcategory, percentage):
        self.name = name
        self.sets = sets
        self.reps = reps
        self.rm = rm
        self.category = category
        self.subcategory = subcategory
        self.percentage = percentage
        self.weight = None

    def calculate_weight(self):
      self.weight = self.rm * (self.percentage / 100)
      return self.weight

    def calculate_tonnage(self):
      self.tonnage = self.sets * self.reps * self.weight * (self.percentage / 100)

    def calculate_volume(self):
      self.volume = self.sets * self.reps

    # Define __str__ method to specify the string representation of the object
    def __str__(self):
        return (f"Exercise(name={self.name}, reps_min={self.sets}, "
                f"reps_max={self.reps}, rm={self.rm}, "
                f"subcategory={self.subcategory}, percentage={self.percentage})")
```

```python

class WeekRoutine:
    def __init__(self,rm, subcategory_list, reps, names_list, percentages, main_lifts) -> None:
        self.rm = rm
        self.main = 'main'
        self.subcategory_list = subcategory_list
        self.reps = reps
        self.names_list = names_list
        self.percentages = percentages
        self.main_lifts = main_lifts

    def getExercises(self):
        exercise_objects = []
        for j in range(len(self.names_list)):
            for i in range(len(self.names_list[j])):
                category = self.main if self.names_list[j][i] in self.main_lifts else None
                exercise = Exercise(
                    self.names_list[j][i],
                    self.reps[i][0],
                    self.reps[i][1],
                    self.rm[j],
                    category,
                    self.subcategory_list[j],
                    self.percentages[i]
                    )
                exercise_objects.append(exercise)
                return exercise_objects




```

```python

class DataFrame:
    def __init__(self, exercise_array, day_names) -> None:
        self.exercise_array = exercise_array
        self.df = pd.DataFrame([vars(e) for e in self.exercise_array])
        self.day_names = day_names
        self.styleFormat()
    def styleFormat(self):
        self.df = self.df.fillna('-')
        self.df['weight'] = [f"{round(ex.calculate_weight(), 1)} kg" for ex in self.exercise_array]
        self.df['percentage'] = self.df['percentage'].astype(str) + '%'
        pd.set_option('display.max_colwidth', 200)




    def defineChunks(self):
        chunk_array = []
        for i in range(0, len(self.df), 4):
            chunk = self.df.iloc[i:i+4]
            chunk = chunk.style.set_caption(f'{self.day_names[i//4]}')
            chunk_array.append(chunk)
            return chunk_array
```
